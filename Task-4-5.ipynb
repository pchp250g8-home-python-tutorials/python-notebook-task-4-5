Сколько воды нужно выпарить из 500 кг. целлюлозной массы, содержащей 85% воды, 
чтобы получить раствор с 75% воды.

```python
mr1 = 500
mh1 = mr1 * 85 /100
mc1 = mr1 - mh1
mr2 = mc1 * 100 / (100 - 75)
mh2 = mr1 - mr2
ans = mh2
print(ans)
```

```python
from sympy.abc import x
from sympy import Eq
from sympy import solve

ans = solve(Eq(500 / x, (100 - 75) / (100 - 85)), x)
ans = 500 * 85 / 100 - ans.pop() * 75 / 100
print(ans)
```

```python
from sympy.abc import x, y
from sympy import Eq
from sympy import solve

ans1 = solve([Eq(x + y, 500), Eq((x * 100) / (x + y), 85)], [x, y])
ans2 = solve(Eq(ans1[y] / (x + ans1[y]) * 100, 25), x)
print(ans1)
print(ans1[x] - ans2.pop())
```
